# **Importing libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# **Importing the dataset**

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/SantiagoMorenoV/50-Startups-Multiple-Reg/main/50_Startups.csv?token=GHSAT0AAAAAACAGCYI6ETRXB54FSR4FVYXYZBHL4OQ')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
dataset.head()
```

# **Descriptive Statistics**

```python
print("\n\033[1m\033[36m\033[6m{:^50}\033[0m".format("Descriptive Statistics"))
print(dataset.describe())
```

# **Distributions and Boxplots**

## **Histograms**

```python
# Selecting our variables
variables = ["R&D Spend", "Administration", "Marketing Spend", "Profit"]

# Creating histograms
for var in variables: 
  sns.histplot(data = dataset, x = var)
  plt.title("Histogram of {}".format(var))
  plt.show()
```

# **Boxplots**

```python
# Creting Boxplots
for var in variables: 
  sns.catplot(data = dataset, x = "State", y = var, kind = "box")
  plt.title("{}'s Boxplot per State".format(var))
  plt.show()
```

# **Violin plots**

```python
# Creting Violin plots
for var in variables: 
  sns.catplot(data = dataset, x = "State", y = var, kind = "violin")
  plt.title("{}'s Violin Plot per State".format(var))
  plt.show()
```
